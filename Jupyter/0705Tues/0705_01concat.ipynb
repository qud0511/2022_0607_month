## DF객체 연결하기
- 여러 개의 File에서 읽은 DF객
- 메서드 실행결과 DF객체
- 여러개의 DF객체를 연결시켜

```python
import pandas as pd
# https://pandas.pydata.org/docs/reference/api/pandas.concat.html?highlight=concat
```

```python
# concat(데이터프레임 리스트, axis=(), ignore_index=False, join='outer', keys=[])
# 데이터 연결하기, 속성형과 형태가 같아야 함.
```

```python
DIR_PATH='../Data/'
```

```python
cc1=pd.read_csv(DIR_PATH + 'concat_1.csv')
cc2=pd.read_csv(DIR_PATH + 'concat_2.csv')
cc3=pd.read_csv(DIR_PATH + 'concat_3.csv')
```

```python
# DF 객체 연결하기 => join='outer'
# 행방향 연결, 기본 인덱스 유지, 컬럼 합집합
```

```python
ccAll=pd.concat([cc1,cc2,cc3]) # verfiy ~ 중복확인?
ccAll
```

```python
ccAll.info()
```

```python
# DF 객체 연결하기 (2) => join='inner'
# 행방향 연결, 기본 인덱스 유지, 컬럼 교집합
```

```python
ccAll2=pd.concat([cc1,cc2,cc3], join='inner')
ccAll2
```

```python
# 기본값이 axis=0 행 방향이다.
```

```python
# DF 객체 연결하기 (3) => ignore_index=True
# 행방향 연결, 기본 인덱스 유지 X, 컬럼 교집합
ccAll3=pd.concat([cc1,cc2,cc3], join='inner', ignore_index=True)
ccAll3
```

```python
# DF객체를 열방향으로 연결
ccColALL=pd.concat([cc1,cc2,cc3], axis=1)
ccColALL
```
