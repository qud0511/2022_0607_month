```python
# 모듈 로딩
import matplotlib.pyplot as plt
```

```python
# 그래프 그리기 => plot( 기본값 y_data) => plot([x_data], y_data, 'r-')
plt.plot([1,2,3,4],[11,22,33,100], 'r--')
plt.plot([1,2,3,4],[1,2,3,4], 'g-', [22,33,30,60,10])
plt.ylabel('TEST DATA')        # y축 라벨
plt.xlabel('Value아')            # x축 라벨
plt.title('TEST MAPPLOT')      # 그래프 타이틀
plt.legend('a')                # 범례
plt.show()
```
