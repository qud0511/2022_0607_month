### apply(함수명, axis=방향/축) 실습

- 원소/요소의 함수 적용

- 입력 : Series의 요소 또는 DateFrame의 Series(즉, 행 또는 열)

```python
import numpy as np
```

```python
# [기본] axis=0 ----- row의 컬럼(열)마다 함수 적용
subDF.apply(np.sum, axis=0)
```

```python
# axis=1 ----- 컬럼의 로우(행)마다 함수 적용
subDF.apply(np.sum, axis=1)
```

```python
# 사용자 정의 함수
def addValue(x):
    print(f'[DEBUG]------------ {x}')
    return x+x

def addTwoValue(x, y):
    print(f'[DEBUG]------------ {x}, {y}')
    return x+y
```

```python
subDF.apply(addValue)
```

```python
subDF.apply(addValue, axis=1)
```

```python
# Series.apply()
subDF['Age'].apply(addValue)
```

```python
# subDF['Age'].apply(addValue, axis=1)
# Series는 2차원이기 때문
```

```python
subDF.apply(addTwoValue, y=2000)
```

```python
# 사용자 정의 함수 -------- 미싱 밸류 찾는 함수
def missingValueCheck(sr):
    print(f'[DEBUG]------ {sr}')
    return sr.isnull()
```

```python
subDF.apply(missingValueCheck)
```

```python
def missingValueSum(sr):
    print(f'[DEBUG]----- {sr}')
    return sr.isnull.sum()
```

```python
import numpy as np
```

```python
subDF.applymap(np.sum)
# apply은 Series 단위마다
# applymap은 원소 하나하나 마다
```

```python
subDF.applymap(addValue)
```

```python
subDF.pipe(addValue)
# 덩어리, 데이터프레임 전체가 들어감
```
