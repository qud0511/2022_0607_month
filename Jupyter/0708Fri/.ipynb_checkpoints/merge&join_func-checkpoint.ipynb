```python
# 모듈 로딩
import pandas as pd
```

```python
# 파일 관련 변수들
DIR='../Data/'
FILE_1=DIR + 'stock price.xlsx'
FILE_2=DIR + 'stock valuation.xlsx'
```

```python
# 데이터 로딩
df1=pd.read_excel(FILE_1)
df2=pd.read_excel(FILE_2)
```

```python
df1
```

```python
df2
```

# DF 합치기 => merge

### [기본] DF 사이의 공통된 컬럼명을 기준으로 교집합을 합친다.


```python
allDF=pd.merge(df1, df2)
allDF
# FILE_1, FILE_2의 경우 id를 기준으로 합쳐짐
```

#### [수정] 2개 DF의 모든 공통 컬럼 기준으로 모든 데이터 합치기(outer)

```python
allDF2=pd.merge(df1, df2, how='outer')
allDF2
```

#### [수정] 2개의 DF에서 왼쪽 기준으로 데이터 합치기Left Outer Join
#### [수정] 2개의 DF에서 오른쪽 기준으로 데이터 합치기Left Outer Join

```python
allDF3=pd.merge(df1, df2, how='left')
allDF3
#  allDF4=pd.merge(df1, df2, how='right')
```

```python
# 컬럼명을 기준으로 합치고 싶을 때, 컬럼명을 지정하게 되면 기존에 기준이 되었던 id는 두개가
# 되고 이를 구분하기 위해 id_X, id_y로 표기됨.
allDF3=pd.merge(df1, df2, how='left', left_on='stock_name', right_on='name')
allDF3
# MergeError: Must pass "right_on" OR "right_index"
```

# DataFrame 메서드 join

```python
a = df1.join(df2, lsuffix='_left', rsuffix='_right')
a
```

```python
# 곽성화
df1.join(df2.set_index('id'). on='id')
```

```python
df1.join(df2.set_index('id'). on='id', how='inner')
```

```python
# 신민수
df1.set_index('id').join(df2.set_index('id'), how='inner')
```

```python
df1.set_index('id').join(df2.set_index('id'), how='outer')
```
