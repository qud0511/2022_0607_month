- CSV 파일을 읽고 컬럼별 최고값과 값의 개수 출력

```python
# 판다스 불러오기
import pandas as pd

# 파일 경로 및 이름 설정
DIR_PATH = '../Data/'
FILE_NAME = DIR_PATH + 'iris2.csv'

# 파일 데이터프레임화
# 데이터에 앞부분이 없으니 header 로 설정을 해줘야 한다.
# colunms 으로 header를 지정할 수 도 있습니다.
df_iris = pd.read_csv(FILE_NAME, header=None, na_values=0)

# 데이터 확인
df_iris
```

```python
df_iris.columns
len(df_iris.columns)
```

```python
# 컬럼별 최고값
for i in range(0, len(df_iris.columns)) :
    print(f'{i}컬럼의 최고값 :', df_iris[i].max())
```

```python
# 값의 갯수
for i in range(0, len(df_iris.columns)) :
    print(f'{i}컬럼값의 갯수 :\n', df_iris[i].value_counts())
```
