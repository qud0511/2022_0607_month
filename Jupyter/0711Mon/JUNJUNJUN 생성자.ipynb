```python
# 날짜 쪼개기
dateDF=googleDF01['날짜'].str.split('-', expand=True)
dateDF.head(3)
```

```python
# '년', '월', '일' 칼럼명 변경
dateDF.rename(columns={0:'년', 1:'월', 2:'일'}, inplace=True)
```

```python
googleDF01 = pd.concat([googleDF01, dateDF], axis=1)
googleDF01
```

```python
try:
    googleDF01.drop(columns=['날짜'], inplace=True)

except KeyError:
    pass
googleDF01
```
