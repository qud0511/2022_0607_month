```python
# 모듈 로딩
import pandas as pd
import numpy as np
```

```python
# 파일 관련 변수들
DIR='../Data/'
FILE=DIR + 'stock-data.csv'
df=pd.read_csv(FILE)
df.head()
```

```python
df.info()
```

```python
# 실습 stock-data.csv
# 접근자 사용하여 Data 컬럼 분할 (year, month, day)
# 일 기준으로 정렬
df.dtypes
```

```python
DateDF=df.Date.str.split('-', expand=True)
DateDF.head()
```

```python
DateDF.rename(columns={0: 'Year', 1:'Month', 2:'Day'}, inplace=True)
```

```python
DateDF
```

```python
df=pd.concat([df, DateDF], axis=1)
df
```
