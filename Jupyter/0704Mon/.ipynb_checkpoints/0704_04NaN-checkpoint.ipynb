### 결측치 갯수 파악, 처리방법 1 : 삭제

```python
df.isnull().sum()
```

### NaN값을 삭제할 것인지 다른 임의의 값으로 채울 것인지 생각해라,
### 원래 자료를 보고 row를 삭제할 것인지, columns을 지울 것인지 고려해라.
###### https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html?highlight=dropna#pandas.DataFrame.dropna

```python
# 2010년도 1월의 최고 온도 - 최저 온도 출력
oneMonth=df.iloc[[0,1]]
oneMonth
```

```python
# NaN값 제거 -> dropna() : 기보값 => 행(row)방향, 1개라도 NaN있으면 삭제
oneMonth.dropna()
```

```python
# 그래서 => 열(columns) 방향 설정, 최고/최저온도값이 모두 NaN인 것을 삭제하 싶음.
# => dropna(axis=1, how='all')
oneMonth.dropna(axis=1, how='all')
```

```python
# 2월의 최고/최저온도 출력
twoMonth=df.iloc[[2,3]].dropna(axis=1, how='all')
twoMonth
```
