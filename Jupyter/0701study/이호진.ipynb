```python
# 모듈 로딩 ----------------------------------------------------------
import pandas as pd

# 파일 관련 변수들 선언 ----------------------------------------------
DIR_PATH='../Data/'
FIL_NAME=DIR_PATH+'iris.csv'
irisDF=pd.read_csv(FIL_NAME)
```

```python
# 1. Length 데이터만 추출해서 출력해주세요.
irisDF.filter(regex='Length')
```

```python
# 2. SepalLength 컬럼에서 길이가 5.0보다 큰 컬럼의 SepalLength와 Names만 출력해주세요
#irisDF[irisDF.SepalLength>5.0]
result=irisDF[irisDF.SepalLength >= 5.0]
result[['SepalLength', 'Name']]
```
