# 남북한발전전력량.xlsx

````python
# 모듈 로딩 ----------------------
import pandas as pd
import numpy as np
````

- 1. DataFrame형식으로 데이터를 가져오세요.

````python
nscharge=pd.read_excel('../Data/남북한발전전력량.xlsx') # read_excel로 엑셀파일 불러오기
nscharge=nscharge.drop([0,5]) # 합계 행 삭제
nscharge['전력량 (억㎾h)']=['남한','남한','남한','남한','북한','북한','북한'] # '전력량 (억㎾h)'열의 값을 변경
nscharge.loc[4]=nscharge.loc[4].replace('-',np.NaN) # 4행의 '-'값을 NaN으로 변경
nscharge.loc[8]=nscharge.loc[8].replace('-',np.NaN) # 8행의 '-'값을 NaN으로 변경
nscharge
````

````python
nscharge=nscharge.set_index(['전력량 (억㎾h)','발전 전력별']) # '전력량 (억㎾h)','발전 전력별' 2개열을 인덱스로 저장
nscharge=nscharge.astype('float64') # 남은 열들인 1990~2016은 'float64'타입으로 변환
nscharge # 숫자 값들이 실수로 변한 것을 확인 할 수 있음
````

```
nscharge=pd.concat([nscharge[nscharge.columns[0:2]],nscharge[nscharge.columns[2:].values].astype('float64')],axis=1)
 - nscharge[nscharge.columns[0:2]]는 '전력량 (억㎾h)','발전 전력별'열의 전체 데이터 의미
 - nscharge[nscharge.columns[2:].values].astype('float64') 1990~2016년의 타입을 'object'에서 'float64'로 바꾸고
 - 2개의 데이터를 열결합 시켜서 nscharge에 저장시킴
nscharge.info() # 각 열들의 타입을 확인
```


- 3. 1990~2016 컬럼에서 컬럼별 최대값, 최소값을 출력해주세요.

````python
nscharge.describe().iloc[[7,3]]
# nscharge의 열들이 'float64' 타입이므로
# describe()를 돌리면 사분위수, 최대값 최소값들 많은 값들이 있음
````

````python
pd.DataFrame([nscharge.max().values,nscharge.max().values],
             index=['max','min'],columns=nscharge.columns)
# 최대최소를 출력하는 다른방법
# nscharge.max().values와 nscharge.max().values은 갈 형태로 데이터프레임에 저장
# index옵션으로 행 이름 지정하고 columns옵션으로 열 이름 지정
````

- 4. 1990~2016 컬럼에서 짝수 컬럼 데이터만 출력해주세요.

````python
nscharge[[ i for i in nscharge.columns if int(i)%2==0]]
# 1990~2016열을 정수로 변환하여 2로 나누어 짝수 년도이면 해당 열 출력
````

````python
pd.DataFrame(nscharge.iloc[:].values[:,[i for i in range(len(nscharge.columns)) if i%2==0]],
            index=nscharge.index,columns=nscharge.columns[[i for i in range(len(nscharge.columns)) if i%2==0]])
# 열의 길이를 for문을 이용해 순서대로 반환 후 2로 나누어 짝수이면 칼럼을 출력
# index옵션으로 행 이름 지정하고 columns옵션으로 열 이름 지정
````

- 5. 전력량과 발전 전력별을 하나의 컬럼을 합치고 합친 걸럼을 추가해 주세요.

````python
li=[] #빈리스트
for ind in nscharge.index: # nscharge.index는 멀티 인덱스로 ['남한','수력'] 등으로 저장되어 있음
    li.append(ind[0]+' '+ind[1]) # 문자열 값을 합쳐서 리스트에 추가
nscharge.index=[li] #만든 리스트로 nscharge 행이름을 변경
nscharge=nscharge.reset_index() # 지정된 행을 컬럼으로 다시 리셋시킴
nscharge.rename(columns={'level_0':'발전전력(나라별)'},inplace=True) # 리셋 시킨 열이름인 'level_0'을 '발전전력(나라별)'로 변경
nscharge
````

- 6. 5번에서 생성된 컬럼을 인덱스로 설정해 주세요.

````python
nscharge=nscharge.set_index('발전전력(나라별)') # '발전전력(나라별)'열을 인덱스로 지정
nscharge
````
