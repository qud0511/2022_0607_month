```python
# 과제 2
irisDF = pd.read_csv(DIR_CSV)

# 2-1
print(irisDF[['SepalLength', 'PetalLength']])

# 2-2
result = irisDF[irisDF['SepalLength'] > 5.0]
print(result[['SepalLength', 'Name']])

# 2-3
result = irisDF[irisDF['Name'] == 'Iris-setosa']
print(result.max())
print(result.min())

result = irisDF[irisDF['Name'] == 'Iris-versicolor']
print(result.max())
print(result.min())

result = irisDF[irisDF['Name'] == 'Iris-virginica']
print(result.max())
print(result.min())
```
