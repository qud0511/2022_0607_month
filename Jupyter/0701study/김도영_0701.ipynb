## 3. 1990~2016 컬럼에서 컬럼별 최대값, 최소값을 출력해주세요.

### 방법 1

```python
dfNum=electricDF.drop(['전력량 (억㎾h)','발전 전력별'],axis=1)       # electricDF에서 [전력량 (억㎾h), 발전 전력별] 컬럼을 없앤 후 dfNum에 저장

dfNum0=dfNum.replace('-',0)              # dfNum에서 '-' 을 0 으로 변경 후 dfNum0에 저장

df_describe=dfNum0.describe()          # dfNum0 컬럼의 기술통계 정보를 df_describe에 저장

df_describe
```

```python
# dfNum 에서 '-' 자리에 연도별 칼럼의 평균 값으로 채우기
# dfNum에서 '-'자리에 1990~2016 각 컬럼 평균 값을 채우게 되더라도 1990~2016 컬럼에서 컬럼별 최대값, 최소값을 출력하는 것에는 지장이 없기 때문에 평균을 넣었습니다.
df_mean=df_describe.loc['mean']     # 연도별 컬럼에서 각 평균 값을 추출하여 df_mean에 저장

list=[]
for i in df_mean:
    list.append(i)       # list에 연도별 컬럼에서 각 평균을 순서대로 담는다.
    
list1=list[:-5]        # 남한의 신재생 row(행인덱스 4)에서 '-'으로 된 요소(1960~2011) 채우기 위해 list에서 필요한 부분만 list1에 저장

dfNum.iloc[4,:-5]=list1       # 남한의 신재생 row에 '-'으로 된 요소만 각 연도별 컬럼의 평균 값으로 바꾸기
dfNum.iloc[8]=list       # 북한의 원자력 row에 각 연도별 컬럼의 평균 값으로 바꾸기

df_astype=dfNum.astype(float)     # dfNum을 astype메서드로 실수로 변환 후 df_astype에 저장

df1_describe=df_astype.describe()    # df_astype 컬럼의 기술통계 정보를 df1_describe에 저장

df1_describe.loc[['max','min']]     # 1990~2016 컬럼에서 컬럼별 최대값, 최소값을 출력
```

## 5. 전력량과 발전 전력별을 하나의 컬럼을 합치고 합친 컬럼을 추가해 주세요.

```python
electricCol=electricDF['전력량 (억㎾h)']   # electricDF 에서 전력량 (억㎾h) 컬럼 electricCol에 저장
electricCol[1:5]='남한'     # 전력량 (억㎾h) 컬럼의 첫번째~네번째 요소 '남한'으로 채우기
electricCol[6:9]='북한'     # 전력량 (억㎾h) 컬럼의 여섯번째~여덟번째 요소 '북한'으로 채우기

list=[]
for i in range(9):
    j=electricDF.loc[i,'전력량 (억㎾h)']+' '+electricDF.loc[i,'발전 전력별']
    list.append(j)      # electricDF에서 전력량과 발전 전력별의 모든 행을 각 행끼리 더하여 리스트에 담는다.
    
electricDF['전력량+발전 전력별']=list       # 전력량과 발전 전력별을 하나의 컬럼으로 합친 컬럼을 추가

electricDF
```

# 2. ‘iris.csv’에서 아래 결과를 도출하세요.

```python
# 모듈 로딩 ----------------------------------------------------------------------------
import pandas as pd

# 파일 관련 변수들 선언 ---------------------------------------------------------------
DIR_PATH='../Data/'
FILE_NAME=DIR_PATH+'iris.csv'

# CSV 파일 읽기 ===> DataFrame으로 저장 ----------------------------------------------
irisDF=pd.read_csv(FILE_NAME)

# 데이터 요약 정보 확인 -----------------------------------------------------------------------
irisDF.info()
```

## 1. Length 데이터만 추출해서 출력해주세요.

```python
# Length 데이터만 추출해서 출력
df=irisDF.drop(['SepalWidth','PetalWidth','Name'],axis=1)               # irisDF 에서 ['SepalWidth','PetalWidth','Name'] 칼럼을 없앤 후 df에 저장
df
```

## 2. SepalLength 컬럼에서 길이가 5.0보다 큰 컬럼의 SepalLength와 Name만 출력해주세요.

```python
# SepalLength 컬럼에서 길이가 5.0보다 큰 컬럼 찾기
SepalLengthCol=irisDF['SepalLength']                    # irisDF에서 SepalLength 컬럼을 SepalLengthCol에 저장
result=SepalLengthCol>5.0            # SepalLengthCol에서 길이가 5.0보다 큰 컬럼을 result에 저장
df1=irisDF[result]                
df1
```

```python
# SepalLength 컬럼에서 길이가 5.0보다 큰 컬럼의 SepalLength와 Name만 출력
df2=df1[['SepalLength','Name']]
df2
```

## 3. Iris-setosa, Iris-versicolor, Iris-virginica에서 각각 최대, 최소 컬럼을 출력해 주세요.

```python
# Iris-setosa에서 각각 최대, 최소 컬럼을 출력
NameCol=irisDF['Name']                   # irisDF 에서 Name 컬럼을 NameCol에 저장
setosa=NameCol=='Iris-setosa'                # NameCol의 요소 중 'Iris-setosa' 와 같은지 참/거짓 판별 후 setosa에 저장
df3=irisDF[setosa].describe()        # irisDF에서 setosa 참인 것만 걸러낸 후 컬럼의 기술통계 정보를 df3에 저장
print(df3.loc['max'])
print(df3.loc['min'])
```

```python
# Iris-versicolor에서 각각 최대, 최소 컬럼을 출력
NameCol=irisDF['Name']               # irisDF 에서 Name 컬럼을 NameCol에 저장
versicolor=NameCol=='Iris-versicolor'       # NameCol의 요소 중 'Iris-versicolor' 와 같은지 참/거짓 판별 후 versicolor에 저장
df4=irisDF[versicolor].describe()      # irisDF에서 versicolor 참인 것만 걸러낸 후 컬럼의 기술통계 정보를 df4에 저장
print(df4.loc['max'])
print(df4.loc['min'])        
```

```python
# Iris-virginica에서 각각 최대, 최소 컬럼을 출력
NameCol=irisDF['Name']          # irisDF 에서 Name 컬럼을 NameCol에 저장
virginica=NameCol=='Iris-virginica'       # NameCol의 요소 중 'Iris-virginica' 와 같은지 참/거짓 판별 후 virginica에 저장
df5=irisDF[virginica].describe()        # irisDF에서 virginica 참인 것만 걸러낸 후 컬럼의 기술통계 정보를 df5에 저장
print(df5.loc['max'])
print(df5.loc['min'])
```
