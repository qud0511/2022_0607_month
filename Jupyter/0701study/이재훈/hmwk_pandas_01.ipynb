### 5. 전력량과 발전 전력별을 하나의 컬럼으로 합치고 새로운 컬럼으로 추가해 주세요.

### 6. 5번에서 생선된 컬럼을 인덱스로 설정해 주세요

```python
# padnas 호출
import pandas as pd
```

```python
# 1. DataFrame 형식으로 데이터를 가져오세요

# '남북한발전전력량.xlsx' 파일 읽어오기, NaN 값을 탐지하지 않음 설정
df_EP = pd.read_excel('../../Data/남북한발전전력량.xlsx', na_values='-')
df_EP.fillna
df_EP
```

```python
# 2. 요약정보를 출력해주세요
print(f'간략정보 : \n{df_EP.info()}')

print(f'행정보 : \n{df_EP.index}', f'열정보 : \n{df_EP.columns}', f'모양정보 : \n{df_EP.shape}', sep='\n\n')
```

```python
# 3. 1990 ~ 2016 컬럼에서 컬럼별 최대값, 최소값을 출력해주세요.

# 1990~2016 까지 컬럼값 추출
year = df_EP.columns[2:]

def yearMaxMin():
    for i in range(0,len(year)):
        df_EP_year = df_EP[year[i]]
        df_EP_year_max = df_EP_year.max()
        df_EP_year_min = df_EP_year.min()
        print(f'{year[i]}의 최대값 : {df_EP_year_max}', f'{year[i]}의 최소값 : {df_EP_year_min}', sep='\n')
        
yearMaxMin()
```

```python
# 4. 1990 ~ 2016 컬럼에서 짝수 컬럼 데이터만 출력해주세요.

def yearEven():
    # year 리스트에서
    for i in range(0,len(year)):
        # 2로 나눠 떨어지는 경우
        if int(year[i]) % 2 == 0 :
            # 해당 연도 컬럼 
            print(f'{year[i]} : \n{df_EP[year[i]]}')
        
yearEven()
```

```python
# 5. 전력량과 발전 전력별을 하나의 컬럼으로 합치고 새로운 컬럼으로 추가해 주세요.
volume=df_EP['전력량 (억㎾h)']
volume[:5]='남한'
volume[5:]='북한'

df_EP['전력량 (억㎾h)_발전 전력별']=df_EP['전력량 (억㎾h)']+'_'+df_EP['발전 전력별']
df_EP
```

```python
# 6. 5번에서 생선된 컬럼을 인덱스로 설정해 주세요
df_EP.index = df_EP['전력량 (억㎾h)_발전 전력별']
df_EP
```
