### 1. DataFrame 형식으로 데이터를 가져오세요

```python
import pandas as pd
FILE_PATH = './data/'
FILE_NAME = '남북한발전전력량.xlsx'
FILE = FILE_PATH + FILE_NAME
snkorea = pd.read_excel(FILE, na_values='-')
```

### 2. 요약정보를 출력해주세요

```python
print(snkorea.info())
```

```python
print(snkorea.index)
print(snkorea.columns)
print(snkorea.shape)
```

### 3. 1990 ~ 2016 컬럼에서 컬럼별 최대값, 최소값을 출력해주세요

```python
sn9016 = snkorea.loc[:,'1990':'2016']
print(sn9016.max())
```

```python
print(sn9016.min())
```

### 4. 1990 ~ 2016 컬럼에서 짝수 컬럼 데이터만 출력해주세요

```python
print(sn9016.loc[:,::2])
```


### 5. 전력량과 발전 전력별을 하나의 컬럼을 합치고 합친 컬럼을 추가해 주세요

```python
print(snkorea.columns)
data1 = snkorea['전력량 (억㎾h)'].fillna(method='ffill')
data2 = snkorea['발전 전력별']
data = data1 + data2
snkorea['남북한총전력'] = data
snkorea
```
