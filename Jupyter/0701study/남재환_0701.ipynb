```python
# 모듈 로딩 ------------------------------------------------------------------------
import pandas as pd

# 파일 관련 변수들 선언 ------------------------------------------------------------
DIR_PATH='../Data/'
FILE_NAME=DIR_PATH+'남북한발전전력량.xlsx'
```

```python
koreaDF=pd.read_excel(FILE_NAME,na_filter=False)
koreaDF.info()
```

```python
koreaDF=koreaDF.transpose()
koreaDF
```

```python
koreaDF.iloc[:]
```

```python
koreaDF.describe()
```

```python
koreaDF.max()
```

```python
koreaDF.loc[:, ::2]
```

```python
koreaDF1=koreaDF.transpose()
koreaDF1
```

```python
result2=koreaDF1['발전 전력별']
result2
```

```python
result3=koreaDF1['전력량 (억㎾h)']
result3
```

```python
result1=result2+result3
result1
```

```python
result1
```

```python
koreaDF1['전력량 (억㎾h)'+'발전 전력별']=result1
koreaDF1
```

```python
koreaDF1.drop(['발전 전력별','전력량 (억㎾h)'],axis=1,inplace=True)
koreaDF1
```

```python
koreaDF2=koreaDF1.set_index('전력량 (억㎾h)'+'발전 전력별')
koreaDF2
```
