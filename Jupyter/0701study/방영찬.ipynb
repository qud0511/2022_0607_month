# 1. ‘남북한발전전력량.xlsx’에서 아래 결과를 도출하세요.

## 1. DataFrame형식으로 데이터를 가져오세요.

```python
import pandas as pd

DIR_PATH='../../Data/'

FILE_NAME= DIR_PATH +'남북한발전전력량.xlsx'

elecDF=pd.read_excel(FILE_NAME)

elecDF
```

## 2. 요약정보 출력해주세요.

```python
elecDF.info()
```

## 3. 1990~2016 컬럼에서 컬럼별 최대값, 최소값을 출력해주세요.

```python
elecDF.replace('-', None)

elecDF_1=elecDF.drop(['전력량 (억㎾h)','발전 전력별'] ,axis=1)
elecDF_1=elecDF_1.replace('-', None)

elecDF_1.max()
```

```python
elecDF_1.min()
```

## 4. 1990~2016 컬럼에서 짝수 컬럼 데이터만 출력해주세요.

```python
elecDF_2=elecDF_1.iloc[:,::2]
elecDF_2
```

## 5. 전력량과 발전 전력별을 하나의 컬럼을 합치고 합친 걸럼을 추가해 주세요.

```python
elecDF

elecDF['전력량 (억㎾h) + 발전 전력별'] = elecDF['전력량 (억㎾h)'] + '+' + elecDF['발전 전력별']
elecDF['전력량 (억㎾h) + 발전 전력별']
```

```python
elecDF_m = pd.merge(elecDF['전력량 (억㎾h) + 발전 전력별'], elecDF)
elecDF_m
```

## 6. 5번에서 생성된 컬럼을 인덱스로 설정해 주세요.

```python
elecDF_m.set_index('전력량 (억㎾h) + 발전 전력별', inplace=True)
```

```python
elecDF_m
```
