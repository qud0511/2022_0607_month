# 4. 1990~2016 컬럼에서 짝수 컬럼 데이터만 출력해주세요.

```python
a = eleDF.iloc[:,2::2]
a
```

```python
# for문을 쓴다면 이렇게 

odd_col = []
for x in a.columns:
    if int(x) % 2 ==0:
        odd_col.append(x)

a[odd_col]
```
