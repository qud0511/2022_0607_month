## 성별, 등석, 나이에 따른 생존율

````python
# 나이가 결측치인 행 삭제
df1=df[df['age'].notnull()]
df1.index=range(len(df1))
df1.info()
````

````python
# 나이를 연령대로 바꾸기
for id in range(len(df1['age'])):
    a=df1.loc[id,'age']
    if a<10:
        df1.loc[id,'age']=0
    elif a<20:
        df1.loc[id,'age']=10
    elif a<30:
        df1.loc[id,'age']=20
    elif a<40:
        df1.loc[id,'age']=30
    elif a<50:
        df1.loc[id,'age']=40
    elif a<60:
        df1.loc[id,'age']=50
    elif a<70:
        df1.loc[id,'age']=60
    elif a<80:
        df1.loc[id,'age']=70
    elif a<90:
        df1.loc[id,'age']=80
    elif a<100:
        df1.loc[id,'age']=90
````
