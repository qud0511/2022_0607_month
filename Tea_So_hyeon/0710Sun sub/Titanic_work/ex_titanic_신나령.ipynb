#### 성별, 등석, 나이에 따른 생존율

```python
# 나이 NaN => 평균 값으로 채우기
df.age.fillna(df.age.mean(),inplace=True)
df.info()
```

```python
# 나이 int로 형변환
df.age.unique()
df.age=df.age.astype(int)
df.age.unique()
```

```python
# 나이 최대, 최소값 확인
df.age.max(), df.age.min()
```

```python
ranges=[-0.9,9.9,19.9,29.9,39.9,49.9,59.9,69.9,79.9,89.9] 
labels=['어린이','10대','20대','30대','40대','50대','60대','70대','80대'] # 위에서 확인한 나이 최대, 최소값으로 범주 설정
age_bins=pd.cut(df.age,ranges,labels=labels)
df.insert(3,'age label',age_bins)
df
```

```python
# 전체 생존율
df.survived.mean()
```

```python
# 성별 생존율
# 하나하나 빼기
sexGr=df.groupby('sex')
rfemale=sexGr.get_group('female').survived.mean()
rmale=sexGr.get_group('male').survived.mean()
pd.DataFrame([rfemale,rmale],index=['female','male'],columns=['survived rate'])
```

```python
# 간단하게
pd.DataFrame(df.groupby('sex').survived.mean())
```

```python
# 나이대 생존율
pd.DataFrame(df.groupby('age label').survived.mean())
```

```python
# 등석 생존율
pd.DataFrame(df.groupby('class').survived.mean())
```

```python
pd.DataFrame(df.groupby(['sex','class','age label']).survived.mean())
```
