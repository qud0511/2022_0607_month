# 성별, 좌석, 나이에 따른 생존율

```python
df.age.isnull().sum()
```

```python
df.sort_values('age', ascending=False,na_position='first')
```

```python
df1=df.loc[:,['survived','pclass','sex','age']]
```

```python
df1.info()
```

```python
df1
```

```python
df2=df1.dropna(axis=0)
df2.info()
```

```python
df2['age type']=(df2['age']//10)*10
df2
```

```python
df2.info()
```

```python
PG=df2.groupby(by='pclass')
SG=df2.groupby(by='sex')
AG=df2.groupby(by='age type')
```

```python
PG.describe()
```

```python
def aggG(x):
    x=x.agg({'survived':'sum', 'sex':'count'})
    x['survived ratio [%]']=x['survived']/x['sex']*100
    x=pd.DataFrame(x)
    print(x)
```

```python
print(f'연령별 생존율 -----------------------------')
aggG(AG)
print(f'좌석별 생존율 ---------------------------')
aggG(PG)
print(f'성별 생존율 -----------------------------')
aggG(SG)
```
