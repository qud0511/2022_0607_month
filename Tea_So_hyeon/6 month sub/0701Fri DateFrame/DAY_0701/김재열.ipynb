# 남북한발전전략량.xlsx에서 아래 결과 도출하세요

1. DataFrame형식으로 데이터를 가져오세요
2. 요약정보 출력해주세요
3. 1990~2016 컬럼에서 컬럼별 최대값, 최소값을 출력해주세요
4. 1990~2016 컬럼에서 짝수 컬럼 데이터만 출력해주세요
5. 전력량과 발전 전력별을 하나의 컬럼을 합치고 합친 컬럼을 추가해주세요
6. 5번에서 생성된 컬럼을 인덱스로 설정해 주세요.

```python
import pandas as pd

DIR_PATH='../Data/'
File_Name=DIR_PATH+'남북한발전전력량.xlsx'
```

```python
# 1. DataFrame형식으로 데이터를 가져오세요
ElecDF=pd.read_excel(File_Name)
ElecDF
```

```python
# 2. 요약정보 출력해주세요
ElecDF.info()
```

```python
# 3. 1990~2016 컬럼에서 컬럼별 최대값, 최소값을 출력해주세요
a=ElecDF.replace('-',None)
a
```

```python
MaxMin=pd.DataFrame([a.iloc[:,2:].max(),a.iloc[:,2:].min()])
MaxMin.rename(index={0:'최대값',1:'최소값'})
```

```python
# 4. 1990~2016 컬럼에서 짝수 컬럼 데이터만 출력해주세요
a.iloc[:,2::2]
```

```python
# 5. 전력량과 발전 전력별을 하나의 컬럼을 합치고 합친 컬럼을 추가해주세요
b=ElecDF.fillna('')
merCol=b['전력량 (억㎾h)']+ a['발전 전력별']
b['발전전력량']=merCol
b
```

```python
# 6. 5번에서 생성된 컬럼을 인덱스로 설정해 주세요.
c=b.set_index('발전전력량')
c
```

# 과제2_ iris.csv에서 결과 도출하기

1. Length 데이터만 추출해서 출력
2. SepalLength 컬럼에서 길이가 5.0보다 큰 컬럼의 SepalLength와 Names만 출력
3. Iris-setosa, Iris-versicolor, Iris-virginica에서 각각 최대, 최소 칼럼 출력


```python
DIR_PATH2='..Data//'
File_Name2=DIR_PATH+'iris.csv'
```

```python
irisDF=pd.read_csv(File_Name2)
irisDF
```

```python
# 1. Length 데이터만 추출해서 출력
pd.DataFrame(irisDF['PetalLength'])
```

```python
# 2. SepalLength 컬럼에서 길이가 5.0보다 큰 컬럼의 SepalLength와 Names만 출력
up5=irisDF[irisDF['SepalLength']>5]
up5
```

```python
up5[['SepalLength','Name']]
```

```python
# 3. Iris-setosa, Iris-versicolor, Iris-virginica에서 각각 최대, 최소 칼럼 출력
setosa=irisDF[irisDF['Name']=='Iris-setosa']
print(setosa.max(),setosa.max())
```

```python
versicolor=irisDF[irisDF['Name']=='Iris-setosa']
print(versicolor.max(),versicolor.max())
```

```python
virginica=irisDF[irisDF['Name']=='Iris-virginica']
print(virginica.max(),virginica.max())
```
