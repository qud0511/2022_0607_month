### 1-3-1. 2개의 컬럼 통합 후 인덱스 설정
'전력량 (억㎾h)'과 '발전 전력별' 컬럼을 '발전 전력별 전력량(억㎾h)'으로 통합 후 인덱스로 설정

```python
# 데이터가 어떤 카테고리로 구성되어 있는지 확인
electDF2 = electDF.copy()
electDF2.columns
```

```python
# '전력량 (억㎾h)'과 '발전 전력별' 컬럼을 합쳐서 첫번째 열에 추가
new = electDF.fillna('').loc[:, '전력량 (억㎾h)']\
      + ' ' + electDF.loc[:, '발전 전력별']
electDF2.insert(loc=0, column='발전 전력별 전력량(억kWh)', value=new)
```

```python
# 기존 컬럼 삭제
electDF2.drop(columns=['전력량 (억㎾h)', '발전 전력별'], inplace=True)
```

```python
# 인덱스 설정
electDF2_fin = electDF2.set_index('발전 전력별 전력량(억kWh)')
electDF2_fin
```

### 1-3-3. 국가, 년도별 전력량 최대, 최소 분야 및 사용량

- 남한과 북한의 전력량 최대, 최소 분야를 각 년도별로 조회
- 전력 사용량 최대값을 구할 때 '합계' 행의 데이터가 집계되지 않도록 데이터를 재구성

```python
# 남한, 북한의 합계 행을 제외한 데이터
electDF3 = electDF2.copy()
electDF3.drop([0, 5], inplace=True)
electDF3
```

```python
# '-'을 제외한 최대값 구하기
year_data = electDF3.loc[:, '1990':'2016']
year_data = year_data[~year_data.isin(['-'])]
m = year_data['1990'].max(skipna=True)
m
```

```python
# 최대값에 해당하는 발전 분야
i = year_data['1990'].index[year_data['1990']==m][0]
electDF3['발전 전력별 전력량(억kWh)'][i]
```

```python
# 남한 데이터
남한 = year_data.iloc[:4]
남한
```

```python
# 북한 데이터
북한 = year_data.iloc[4:6]
북한
```

```python
def test_max(nation):
    for year in electDF3.columns[1:]:
        m = nation[year].max(skipna=True)
        i = nation[year][nation[year]==m].index[0]
        print(year+"년 전력량 사용 최대 분야:"
              +electDF3['발전 전력별 전력량(억kWh)'][i]
              +" "+str(m)+"(억kWh)")
print('남한 년도별 전력량 사용 최대 분야 및 사용량')
test_max(남한)
```

```python
def test_min(nation):
    for year in electDF3.columns[1:]:
        m = nation[year].min(skipna=True)
        i = nation[year][nation[year]==m].index[0]
        print(year+"년 전력량 사용 최소 분야:"
              +electDF3['발전 전력별 전력량(억kWh)'][i]
              +" "+str(m)+"(억kWh)")
print('남한 년도별 전력량 사용 최소 분야 및 사용량')
test_min(남한)
```

```python
print('북한 년도별 전력량 사용 최대 분야 및 사용량')
test_max(북한)
```

```python
print('북한 년도별 전력량 사용 최소 분야 및 사용량')
test_min(북한)
```

### 1-3-4. 국가, 전력별 전력량 최대, 최소 년도

남한과 북한의 전력별 전력량이 가장 많았던 년도와 가장 적었던 년도

```python
year_data2 = electDF3[~electDF3.isin(['-'])]
field = list(electDF3['발전 전력별 전력량(억kWh)'][:4])
```

```python
def test_max2(nation):
    n = 0
    for row in nation.index:
        m = year_data.loc[row].max(skipna=True)
        i = year_data.loc[row][year_data.loc[row]==m].index[0]
        print(field[n]+' 발전 전력 사용 최대 년도: '+str(i)+'년, '
              +'사용량: '+str(m)+'(억kWh)')
        n += 1
print('남한 전력별 사용량 최대 년도 및 사용량')
test_max2(남한)
```
