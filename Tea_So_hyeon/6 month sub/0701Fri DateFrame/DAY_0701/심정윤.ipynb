## 3. 1990~2016 컬럼에서 컬럼별 최대값, 최소값을 출력해주세요.



```python
# 제일 앞 두 열 삭제
sn_DF2=sn_DF.drop(['전력량 (억㎾h)','발전 전력별'],axis=1)



for i in sn_DF2.columns:
    print(f'{i} -> [ min : {sn_DF[i].min()} ], [ Max : {sn_DF[i].max()} ]')


```

## 4. 1990~2016 컬럼에서 짝수 컬럼 데이터만 출력해주세요.

```python
DF=pd.DataFrame()
DF.index=[]
for i in sn_DF2.columns:
    if int(i)%2==0:
        DF[i]=sn_DF2[i]
DF
        
```

## 5. 전력량과 발전 전력별을 하나의 컬럼을 합치고 합친 컬럼을 추가해 주세요.

```python
volume=sn_DF['전력량 (억㎾h)']
volume[1:5]='남한'
volume[5:]='북한'
sn_DF
```

```python
result=sn_DF['전력량 (억㎾h)']+'_'+sn_DF['발전 전력별']
result
```

## 6. 5번에서 생성된 컬럼을 인덱스로 설정해 주세요.

```python
sn_DF['Category']=result
sn_DF.index=sn_DF['Category']
sn_DF
```
