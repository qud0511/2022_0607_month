# 1-1. DataFrame형식으로 데이터를 가져오세요.

```python
# 모듈 로딩----------------------------------------------------------
import pandas as pd

# 파일 관련 변수들 선언------------------------------------------------------------------------------
DIR_PATH = '../Data/'
FILE_NAME = DIR_PATH+"남북한발전전력량.xlsx"

# 엑셀 파일 읽기 ==> DataFrame으로 저장 --------------------------------------------------------------
EnergyDF=pd.read_excel(FILE_NAME)
```

# 1-3. 1990~2016 컬럼에서 컬럼별 최대값, 최소값을 출력해주세요.


```python
EnergyDF
```

```python
# 잘못된 데이터 값 변경 ( '-' -> None) : 최대값, 최소값 계산을 위해

EnergyDF1=EnergyDF.replace('-',None)   # 원본을 변경하지 않고, 메인으로 사용할 EnergyDF1 생성
EnergyDF1                               # Energy1 = '-' 을 None 으로 변경한 EnergyDF 
```

```python
# only_yearDF = EnergyDF1에서 연도 컬럼들에 대한 데이터만 추출한 DataFrame

# 방법 1 : 인덱스 슬라이싱 사용. -> .columns[:] 를 사용해야 함.
only_yearDF=EnergyDF1[EnergyDF1.columns[2:]] 

# 방법 2 : .iloc 사용.
only_yearDF=EnergyDF1.iloc[:,2:] # 행 전체, 열[2]부터 끝까지 추출.

print(only_yearDF)
print(f'1990~2016 최대값 :\n{only_yearDF.max()}')   # 연도별 최대값 출력
print(f'1990~2016 최소값 :\n{only_yearDF.min()}')   # 연도별 최소값 출력
```

# 1-4.1990~2016 컬럼에서 짝수 컬럼 데이터만 출력해주세요.

```python
# list : 짝수 컬럼만 리스트에 분류.

# 방법1 : for문 사용
list=[]
for i in only_yearDF.columns:
    if int(i)%2==0: 
        list.append(i)
print(list)    
print(only_yearDF[list]) # 데이터프레임에서 짝수 연도 컬럼만 불러오므로 EnergyDF[list], EnergyDF1[list]도 같은 결과가 나옴.

# 방법2: 슬라이싱 이용
only_yearDF.iloc[:,::2]
```

# 1-5. 전력량과 발전 전력별을 하나의 컬럼을 합치고 합친 걸럼을 추가해 주세요.

```python
EnergyDF1
```

```python
# EnergyDF2 : 결측치에 대당하는 값(None)을 ""으로 변경. 
EnergyDF2=EnergyDF1.fillna("")   

# 2개의 컬럼 데이터를 합친 새로운 컬럼 추가.
EnergyDF2["전력량,발전 전력별"] = EnergyDF2['전력량 (억㎾h)'] + EnergyDF2["발전 전력별"]   
EnergyDF2
```

```python
# 합칠 때 이용한 두개의 컬럼을 삭제. 

# 방법1 : inplace = True 사용할 경우 : EnergyDF2 원본에서 삭제하므로 결과값을 새로운 변수로 받지 않는다.
EnergyDF2.drop(['전력량 (억㎾h)','발전 전력별'],axis=1,inplace=True) 
EnergyDF2

# 방법2 : inplace = True 사용하지 않을 경우 : EnergyDF2 원본은 변하지 않으므로 결과값을 변수로 받아야 한다.
# EnergyDF2 = EnergyDF2.drop(['전력량 (억㎾h)','발전 전력별'],axis=1)  
```

# 2. ‘iris.csv’에서 아래 결과를 도출하세요.

- [데이터 추출]

    - 1. Length 데이터만 추출해서 출력해주세요.

    - 2. SepalLength 컬럼에서 길이가 5.0보다 큰 걸럼의 SepalLength와 Names만 출력해주세요.

    - 3. Iris-setosa, Iris-versicolor, Iris-virginica에서 각각 최대, 최소 컬럼을 출력해 주세요.



```python
# 모듈 로딩----------------------------------------------------------
import pandas as pd

# 파일 관련 변수들 선언------------------------------------------------------------------------------
DIR_PATH = '../Data/'
FILE_NAME = DIR_PATH+'iris.csv'

# CSV 파일 읽기 ==> DataFrame으로 저장 --------------------------------------------------------------
irisDF=pd.read_csv(FILE_NAME)

# 데이터 요약 정보 확인하기
irisDF.info()
```

# 2-1. Length 데이터만 추출해서 출력해주세요.

```python
irisDF
```

```python
# 'Length' 가 포함된 컬럼만 list에 분류.

# 방법1 
list=[]
for i in irisDF.columns:
    if'Length' in i: 
        list.append(i)
print(list)    
irisDF[list]
```

```python
# 방법2 : .filter(regex= ~) 사용
irisDF.filter(regex='Length')
```

# 2-2.SepalLength 컬럼에서 길이가 5.0보다 큰 걸럼의 SepalLength와 Names만 출력해주세요.

```python
irisDF
```

```python
# irisDF1 : 길이가 5.0보다 큰 컬럼에 해당하는 데이터 boolean값.
irisDF1=irisDF['SepalLength'] > 5.0
print(irisDF1)

# irisDF2 : irisDF 전체 데이터에서 boolean 값이 True인 데이터 추출. 
irisDF2=irisDF[irisDF1]
print(irisDF2)

# irisDF2 에서 추출한 데이터 중 'SepalLength','Name' 컬럼에 해당하는 데이터만 출력.
irisDF2[['SepalLength','Name']] 
```

# 2-3. Iris-setosa, Iris-versicolor, Iris-virginica에서 각각 최대, 최소 컬럼을 출력해 주세요.

```python
# result_~ : Name 컬럼의 데이터가 Iris~ 와 같으면 True -> 앞의 boolean 결과값 중 True 인 데이터만을 irisDF에서 출력 
result_setosa=irisDF[irisDF['Name']=='Iris-setosa']
result_versicolor=irisDF[irisDF['Name']=='Iris-versicolor']
result_virginica=irisDF[irisDF['Name']=='Iris-virginica']
```

```python
# 각각 최대, 최소 출력
print(f'Iris-setosa 최대:\n{result_setosa.max()}')
print(f'Iris-setosa 최소:\n{result_setosa.min()}')
print(f'Iris-versicolor 최대:\n{result_versicolor.max()}')
print(f'Iris-versicolor 최소:\n{result_versicolor.min()}')
print(f'Iris-virginica 최대:\n{result_virginica.max()}')
print(f'Iris-virginica 최소:\n{result_virginica.min()}')
```
