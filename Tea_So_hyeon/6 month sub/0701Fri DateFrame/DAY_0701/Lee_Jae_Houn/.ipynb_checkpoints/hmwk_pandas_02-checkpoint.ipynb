### 2. SepalLegth 컬럼에서 길이가 5.0보다 큰 컬럼의 SepalLength와 Name만 출력해주세요.

### 3. Iris-setosa, Iris-versicolor, Iris-virginica에서 각각 최대, 최소 컬럼을 출력해주세요.

```python
# pandas import
import pandas as pd
```

```python
# 경로 설정
DIR_PATH = '../../Data'

# pandas를 이용해 iris.csv 데이터프레임 설정
df_iris = pd.read_csv(f'{DIR_PATH}/iris.csv')

df_iris
```

```python
# 1. Length 데이터만 추출해서 출력해주세요.

# columns 값 추출
df_iris_columns = df_iris.columns

# 처음부터 끝 열까지
for i in range(0, len(df_iris_columns)):
    # 컬럼명에 'Length'라는 단어가 포함되어 있으면
    if 'Length' in df_iris_columns[i]:
            # 해당 컬럼을 프린트
            print(df_iris[df_iris_columns[i]])        
```

```python
# 2. SepalLength 컬럼에서 길이가 5.0보다 큰 컬럼의 SepalLength와 Name만 출력해주세요.

# df_iris_SepalLength에서 길이가 5.0보다 큰 경우
df_iris_SepalLength_5 = df_iris[df_iris['SepalLength'] > 5.0]
# df_iris_SepalLength_5에서 SepalLength과 Name만 출력
df_iris_SepalLength_5[['SepalLength', 'Name']]
```

```python
# 3. Iris-setosa, Iris-versicolor, Iris-virginica에서 각각 최대, 최소 컬럼을 출력해주세요.

#  Name 값을 입력받으면  해당 하는 부분의 최대, 최소를 출력해주는 함수
def defMaxMin(name) :
    # df_iris의 Name 열에 입력받은 값과 같은 행만 추출
    df_iris_flower = df_iris[df_iris['Name'] == name]
    # df_iris_flower의 최대값 추출
    df_iris_flower_max = df_iris_flower.max()
    # df_iris_flower의 최소값 추출
    df_iris_flower_min = df_iris_flower.min()
    # df_iris_flower_max, df_iris_flower_min 출력
    print(f'{name} 최대값 : \n{df_iris_flower_max}', f'{name} 최소값 : \n{df_iris_flower_min}')
    
defMaxMin('Iris-setosa')
defMaxMin('Iris-versicolor')
defMaxMin('Iris-virginica')
```
