# DF 객체연결하기
- 여러개의 File에서 읽은 DF객체
- 메서드 실행 결과 DF객체
- 여러개의 DF객체를 연결 시켜서 하나의 DF객체 생성

```python
# 모듈 로딩
import pandas as pd

# 파일 경로 관련 변수 선언
DIR_PATH = '../Data/'
FILE_NAME1 = 'concat_1.csv'
FILE_NAME2 = 'concat_2.csv'
FILE_NAME3 = 'concat_3.csv'
```

```python
# csv file => dataframe 형태 데이터로 로딩
df1 = pd.read_csv(DIR_PATH + FILE_NAME1)
df2 = pd.read_csv(DIR_PATH + FILE_NAME2)
df3 = pd.read_csv(DIR_PATH + FILE_NAME3)
print(df1,'\n',df2,'\n',df3)
```

```python
# dataframe 요약정보 확인
print(df1.info(),'\n',df2.info(),'\n',df3.info())
```

```python
# dataframe 객체들 연결하기 => pandas.concat([df객체 리스트])
# (1)행방향 연결, 기본 인덱스 유지, 모든 컬럼의 합집합 => join='outer'
df = pd.concat([df1, df2, df3])
# 연결된 dataframe 확인
df
```

```python
# (2)행방향 연결, 기본 인덱스 유지, 모든 컬럼의 교집합 => join='inner'
df_2 = pd.concat([df1, df2, df3], join='inner')
# 연결된 dataframe 확인
df_2
```

```python
df2.index, df2.columns
```

```python
# (3)행방향 연결, 기본 인덱스 유지 X, 모든 컬럼의 교집합 => join='inner'
df_3 = pd.concat([df1, df2, df3], join='inner', ignore_index=True)
# 연결된 dataframe 확인
df_3
```

```python
# df객체를 열방향으로 연결 => concat([df객체 리스트], axis = 1)
df_4 = pd.concat([df1, df2, df3], axis=1)
df_4
```
