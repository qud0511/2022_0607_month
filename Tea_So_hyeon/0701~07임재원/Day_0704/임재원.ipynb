- CSV파일 읽고 컬럼별 최고값과 최고값의 갯수 출력

```python
# 모듈 로딩
import pandas as pd

# 파일 관련 함수 선언
DIR_PATH = r'../Data/'
FILE_NAME = DIR_PATH + 'iris2.csv'
```

```python
# (1) csv => Dataframe으로 가져오기(단, 0번 row는 읽지않는다)
col = ['SL','SW','PL','PW','N']
iris2 = pd.read_csv(FILE_NAME, header=None, names=col)
iris2DF = pd.DataFrame(iris2)
iris2DF
```

```python
max_min_data = pd.DataFrame({'max':iris2DF.max(numeric_only = True),
                             'max_count':[len(iris2DF.loc[iris2DF['SL'] == iris2DF['SL'].max()]),
                                          len(iris2DF.loc[iris2DF['SW'] == iris2DF['SW'].max()]),
                                          len(iris2DF.loc[iris2DF['PL'] == iris2DF['PL'].max()]),
                                          len(iris2DF.loc[iris2DF['PW'] == iris2DF['PW'].max()])],
                             'min':iris2DF.min(numeric_only = True),
                             'min_count':[len(iris2DF.loc[iris2DF['SL'] == iris2DF['SL'].min()]),
                                          len(iris2DF.loc[iris2DF['SW'] == iris2DF['SW'].min()]),
                                          len(iris2DF.loc[iris2DF['PL'] == iris2DF['PL'].min()]),
                                          len(iris2DF.loc[iris2DF['PW'] == iris2DF['PW'].min()])]
                            }).transpose()
max_min_data
```
