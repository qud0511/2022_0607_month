```python
# 모듈 로딩
import pandas as pd

# 파일 관련 변수 선언
DIR = '../Data/'
FILE = DIR + 'country_timeseries.csv'
```

```python
# (1) DF객체 로딩
ebolaDF = pd.read_csv(FILE)
ebolaDF
```

```python
ebolaDF['Date'] = pd.to_datetime(ebolaDF['Date'])
ebolaDF.set_index('Date', inplace=True)
```

```python
# (2) 데이터 확인
ebolaDF.info()
```

```python
ebolaDF.isnull().sum()
```

```python
# (3) 전처리
# NaN 데이터 처리 1. 행 또는 열의 모든 data가 결측치면 drop한다.
ebolaDF.dropna(how='all', inplace=True)
ebolaDF.dropna(how='all', axis=1, inplace=True)
# NaN 데이터 0명으로 치환
ebolaDF.fillna(0, inplace=True)
# Data type 정리
ebolaDF.info()
```

```python
# (4)데이터 가공
# ebola 바이러스 발병 후 몇일이 지났는지에 대한 컬럼 추가
from datetime import datetime as dt

today = dt.today()
print(today)
ebolaDF['Date'] = pd.to_datetime(ebolaDF['Date'])
d_days = today - ebolaDF['Date'].iloc[-1]
```
