### 과제 0704 - (1) 2010년 1~12월까지 최고-최저온도 데이터 추출

```python
# 모듈 로딩
import pandas as pd

# 파일 관련 변수
DIR_PATH='../Data/'
FILE_NAME=DIR_PATH+'weather.csv'
```

```python
# 데이터프레임
wDF=pd.read_csv(FILE_NAME)
```

```python
wDF.info()
```

```python
# 결측치 갯수 파악
wDF.isna().sum()
```

```python
# 누락 데이터 제거 (정상 데이터 1개 미만 시 열 삭제)
wDF.dropna(thresh=1,axis=1)
```

```python
# 'element' 컬럼 데이터명 변경
wDF=wDF.replace('tmax', '최고 온도')
wDF=wDF.replace('tmin', '최저 온도')

# 'month'컬럼과 'element'컬럼 합치기
wDF.month=wDF.month.astype(str)
wDF['월별 온도']=wDF['month']+'월 '+wDF['element']
```

```python
# 합친 컬럼을 인덱스로 설정
wDF=wDF.set_index(['월별 온도'])

# 필요 없는 데이터 삭제 후 행,열 변환
wDF=wDF.drop(['id', 'year', 'month', 'element'], axis=1)
wDF=wDF.T
```

```python
# 월별 최고 온도 출력
highDF=wDF[wDF.columns[::2]].max()
highDF
```

```python
# 월별 최저 온도 출력
lowDF=wDF[wDF.columns[1::2]].min()
lowDF
```
