## 2010년 1월~12월까지 최고온도 - 최저온도 데이터 추출

```python
import pandas as pd
DIR = '../Data/'
NAME = 'weather.csv'
FILE = DIR + NAME
weatherDF = pd.read_csv(FILE)
weatherDF.head()
```

```python
for i in range(0,12,2):
    month = weatherDF.iloc[[i,i+1]]
    month = month.dropna(axis=1, how='all')
    print(month)
```
