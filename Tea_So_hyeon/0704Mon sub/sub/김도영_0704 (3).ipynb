# 2. ‘시도별 전출입 인구수.xlsx’에서 서울에서 다른지역으로 이동한 년도별 인구수데이터 추출 및 분석

```python
# 모듈 로딩
import pandas as pd

# 파일관련 변수 선언
DIR_PATH='../Data/'
FILE_NAME=DIR_PATH+'시도별 전출입 인구수.xlsx'

FILE_NAME
```

```python
# (1) EXCEL FILE => DataFrame으로 로딩
populationDF=pd.read_excel(FILE_NAME)

# (2) 데이터 정보 확인
populationDF.info()
```

```python
populationDF.head()    # 데이터가 잘 들어왔는지 상위 다섯행을 봅니다.
```

```python
populationDF.isnull().sum()    # 결측치 갯수 파악
```

```python
popCol1=populationDF.전출지별     # '전출지별' 컬럼을 popCol1로 저장
```

```python
popCol1.unique()  # '전출지별' 컬럼의 고유값 -> 값의 종류 확인
```

```python
popFill=popCol1.fillna(method='ffill')     # '-' 으로 된 값을 위의 값으로 채운 후 popFill에 저장
populationDF.전출지별=popFill        # popFill을 '전출지별' 컬럼에 다시 저장
populationDF
```

```python
popCol2=populationDF.set_index('전입지별')   # '전입지별' 컬럼을 populationDF 인덱스로 설정하고 popCol2에 저장
allCountry=popCol2.loc['전국']       # popCol2에서 '전국' 행을 allCountry에 저장
all=allCountry.iloc[1]              # allCountry 에서 행인덱스 1 즉, 전출지별 '서울특별시', 전입지별 '전국' 행 을 all에 저장
allNum=all.drop('전출지별')         # all에 '전출지별' 을 없앤 후 allNum에 저장
```

```python
popCol3=populationDF.set_index('전입지별')         # '전입지별' 컬럼을 populationDF 인덱스로 설정하고 popCol3에 저장
seoul=popCol3.loc['서울특별시']            # popCol3에서 '서울특별시' 행을 seoul에 저장
seo=seoul.iloc[1]                           # seoul 에서 행인덱스 1 즉, 전출지별 '서울특별시', 전입지별 '서울특별시' 행을 seo에 저장
seoulNum=seo.drop('전출지별')                  # seo에 '전출지별' 을 없앤 후 seoulNum에 저장
```

```python
print(f'서울에서 다른지역으로 이동한 년도별 인구수데이터 추출 및 분석\n{allNum-seoulNum}')   # '전출지별' '서울특별시' -> 전입지별 '전국'에서 '서울특별시'를 제외한 즉 서울에서 전국에서 서울을 제외한 다른 지역으로 이동한 년도별 인구수데이터 추출 및 분석을 할 수 있습니다.
```

```python
seoulDF=populationDF[populationDF['전출지별']=='서울특별시']      # populationDF 에서 전출지별 컬럼에 '서울특별시'가 들어있는 데이터 프레임을 서울 데이터 프레임으로 만듭니다.
seoulDF=seoulDF.replace('-',0) # seoulDF 의 '-'을 0으로 바꾼다.
```
