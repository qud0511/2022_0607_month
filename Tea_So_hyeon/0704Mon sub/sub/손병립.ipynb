```python
# 모듈 로딩
import pandas as pd

# 파일 관련 변수 선언
DIR_PATH='../Data/'
FILE_NAME=DIR_PATH+'weather.csv'
```

```python
weather=pd.read_csv(FILE_NAME)
weatherD=weather.iloc[:,4:]
```

```python
for i in range(0,21,2):
    weather=weatherD.iloc[i:i+2].dropna(axis=1,how='all')
    m=weather.mean(axis=1)[i]-weather.mean(axis=1)[i+1]
    if i//2+1==9:
        pass
    elif i//2+1>9:
        print(f'{i//2+2}월 : {m}')
    else:
        print(f'{i//2+1}월 : {m}')
```

```python
# 2.시도별 전출입 인구수.xlsx’에서 서울에서 다른지역으로 이동한 년도별 인구수 데이터 추출 및 분석

# 파일 관련 변수 선언
DIR_PATH='../Data/'
FILE_NAME=DIR_PATH+'시도별 전출입 인구수.xlsx'
pop=pd.read_excel(FILE_NAME)
pop.head()
```
