# # 과제 0704-(1) 2010년 1월~12월까지 최고온도-최저온도 데이터 추출

```python
# 모듈 로딩
import pandas as pd

# 파일관련 변수 선언
DIR_PATH='../Data/'
FILE_NAME=DIR_PATH+'weather.csv'

FILE_NAME      # 파일명 맞는지 확인합니다.
```

```python
# (1) CSV FILE => DataFrame으로 로딩
weatherDF=pd.read_csv(FILE_NAME)

# (2) 데이터 정보 확인
weatherDF.info()
```

```python
weatherDF.head()   # 데이터가 잘 들어왔는지 위에 다섯 행을 봅니다.
```

```python
# 결측치 갯수 파악
weatherDF.isnull().sum()
```

```python
print('2010년 1월~12월까지 최고온도-최저온도 데이터 추출''\n')
for i in range(1,9):     #for 문을 사용하여 i는 1부터 8까지
    value=weatherDF.iloc[2*(i-1):2*i].dropna(axis=1, how='all')   # column방향 설정, 최고/최저온도값이 모두 NaN인것 삭제, 남은 데이터를 변수 value에 저장
    result=value.drop(['id','year','month','element'],axis=1).transpose().describe()   # value 에서 숫자 데이터만 남기고 나머지 컬럼 지운 후 행&열 전환하여 컬럼의 기술통계 정보를 변수 result에 저장
    print(f"{i}월의 최고온도 : {result.loc['max'].iloc[0]}")     # result에서 각 월마다 해당하는 max 값을 출력 합니다.    
    print(f"{i}월의 최저온도 : {result.loc['min'].iloc[1]}")      # result에서 각 월마다 해당하는 min 값을 출력 합니다.
                                                                    
for j in range(9,12):
    value=weatherDF.iloc[2*(j-1):2*j].dropna(axis=1, how='all')
    result=value.drop(['id','year','month','element'],axis=1).transpose().describe()
    print(f"{j+1}월의 최고온도 : {result.loc['max'].iloc[0]}")
    print(f"{j+1}월의 최저온도 : {result.loc['min'].iloc[1]}")
```
