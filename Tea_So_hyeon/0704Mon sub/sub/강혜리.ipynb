```python
# 모듈 부팅 ----------------------------------------------------------------
import pandas as pd
DIR_PATH = '../data/'

# 파일 불러오기 -------------------------------------------------------------
FILENAME = DIR_PATH + 'weather.csv'
FILENAME
```

```python
# 파일 읽기 ----------------------------------------------------------------
weatherDF = pd.read_csv(FILENAME)
weatherDF.info()
```

```python
# 파일 정보확인 -------------------------------------------------------------
weatherDF.head(25)
```

```python
# 월별 최대값 도출 -----------------------------------------------------------
# max값이든 행 도출, 컬럼 중 NaN값이 모두 없는 열 삭제, 필요없는 컬럼 삭제
# NaN값 0으로 변경, day컬럼 중 최대값 도출

wM = weatherDF.iloc[::2, 2:].dropna(axis=1, how='all').drop('element', axis=1)
wMDF = wM.set_index(keys=['month'], inplace=False, drop=True).fillna(0).max(axis=1)
```

```python
# 월별 최소값 도출 ------------------------------------------------------------
wN = weatherDF.iloc[1::2, 2:].dropna(axis=1, how='all').drop('element', axis=1)
wNDF = wN.set_index(keys=['month'], inplace=False, drop=True).fillna(100).min(axis=1)
```

```python
# 데이터 프레임 합치기 ---------------------------------------------------------
# 행, 열 도치
# 인덱스 설정

weatherMN = pd.concat([wMDF,wNDF],axis=1).transpose()
weatherMN.index=['최대값','최소값']
pd.DataFrame(weatherMN)
```
