### 1. 결측치(Missing Value) 처리 문제
- 2010년도 1월 ~ 12월 최고온도 – 최저온도 추출

```python
import pandas as pd
```

```python
weather = pd.read_csv('../Data/weather.csv')
```

```python
weather
```

```python
weather.info()
```

```python
month_weather = weather[weather['month'] == 1].dropna(axis = 1, how = 'all')
month_weather['max/ min'] = [month_weather[month_weather.columns[4:]].iloc[0].max(),
                             month_weather[month_weather.columns[4:]].iloc[1].min()]
month_weather.drop(month_weather.columns[4:-1], axis = 1, inplace = True)
month_weather
```

과제 0704 (1) 1월 ~ 12월 까지 최고온도, 최저온도 데이터 추출

```python
year_weather= pd.DataFrame(columns = month_weather.columns)
for m in range(1,13):
    try:
        month_weather = weather[weather['month'] == m].dropna(axis = 1, how = 'all')
        month_weather['max/ min'] = [month_weather[month_weather.columns[4:]].iloc[0].max(),
                                     month_weather[month_weather.columns[4:]].iloc[1].min()]
        month_weather.drop(month_weather.columns[4:-1], axis = 1, inplace = True)
        year_weather = pd.concat([year_weather,month_weather])
    except:
        pass
        
```

```python
year_weather
```
