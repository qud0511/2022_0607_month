## 1. weather.csv파일 활용
    - 2010년 1월에서 12월까지 최고/최저 온도 구하기

````python
# 모듈 로딩
import pandas as pd
import matplotlib as plt
````

````python
# 경로 설정
PATH='../Data/'
FILE_NAME=PATH+'weather.csv'
````

````python
# 파일 읽기
weather=pd.read_csv(FILE_NAME)
weather
````

````python
# 데이터 정보 확인
print(weather.info())
print()
print(weather.index)
print()
print(weather.columns)
print()
print(weather.shape)
````

````python
# 열별 결측치 삭제
we_test=weather.dropna(axis=1,thresh=4)
we_test

# axis=1 => 열별로
# thresh=4 => 각 열 별로 미 결측치 값이 4개 이상이면 그 행은 유지
````

````python
# 데이터 정보 확인
we_test.info()

# id와 element열 말고는 정수나 실수인 숫자형
````

````python
# 최고/최저 온도 데이터 나누기
we_test_hi=we_test[we_test['element']=='tmax']
we_test_lo=we_test[we_test['element']=='tmin']
print(we_test_hi)
print(we_test_lo)

# we_test_hi에는 최고 온도만 있는 데이터 프레임
# we_test_lo에는 최저 온도만 있는 데이터 프레임
````

````python
# 최고/최저 데이터프레임에 결측치 없애기
we_test_hi=we_test_hi.fillna(method='ffill')
we_test_lo=we_test_lo.fillna(method='ffill')

we_test_hi=we_test_hi.fillna(method='bfill')
we_test_lo=we_test_lo.fillna(method='bfill')

print(we_test_hi)
print(we_test_lo)
````

````python
# we_test_hi데이터 프레임에서 각 행의 최고 온도 구하기
ma_ondo=[]
for idx in range(11):
    ma_ondo.append(we_test_hi.iloc[idx,4:].max())

we_test_hi['m_ondo_col']=ma_ondo
we_test_hi

# 빈리스트 생성후 'd2'~'d29'까지의 온도중 max값을 구해서 빈리스트에 추가하기
# 11개의 값이 입력된 리스트를 데이터 프레임에 새로운 열을 만들어 저장하기
````

````python
# we_test_lo데이터 프레임에서 각 행의 최고 온도 구하기
mi_ondo=[]
for idx in range(11):
    mi_ondo.append(we_test_lo.iloc[idx,4:].min())
    
we_test_lo['m_ondo_col']=mi_ondo
we_test_lo

# 빈리스트 생성후 'd2'~'d29'까지의 온도중 min값을 구해서 빈리스트에 추가하기
# 11개의 값이 입력된 리스트를 데이터 프레임에 새로운 열을 만들어 저장하기
````

````python
# 다시 행결합으로 데이터 프레임 합치기
we_test=pd.concat([we_test_hi,we_test_lo])
we_test.sort_index(inplace=True)
we_test

# 결합을 하면 최고온도는 위에 최저 온도는 밑에 있어서
# 인덱스 이름 별로 정리를 하여 다시 월별로 묶어 월별로 보기 쉽게 정렬
# inplace=True로 원본은 변경을 함
````

````python
# 필요 없는 열 삭제 + 인덱스 설정
we_test_fi=we_test[['year','month','element','m_ondo_col']]
we_test_fi.set_index(['year','month','element'],inplace=True)
we_test_fi

# 년+월+최고/최저 온도를 멀티 인덱스로 묶어서 온도 데이터만 값으로 나오게 함
# 2010년 1월의 최고온도는 36.3????? 최저 온도는 10.7이라고 예측함
````

````python
# 전치 행렬
we_test_fi.T
````

````python
# 경로 설정
PATH='../Data/'
FILE_NAME=PATH+'시도별 전출입 인구수.xlsx'
````

````python
# 파일 읽기
popu=pd.read_excel(FILE_NAME)
se_po=popu.copy()[20:37]
se_po['전출지별']='서울특별시'
se_po.index=range(len(se_po))
se_po

# 파일을 읽어오니 열은 년도별로 하는 것이 좋을 것 같아서
# header=0이 기본값으로 첫줄을 열이름로 정하고 
# 서울에서 다른지역으로 이동한 데이터가 필요하니까 전출지별이 서울특별시인 행이
# 19~36행인데 19행은 합계라서 20행부터 36행 까지의 행만 추출하여 새로운 데이터플레임을 만듬
# 인덱스도 0부터 설정을 했고 '전출지별'열의 전체 이름을 서울특별시로 값을 넣음
````
