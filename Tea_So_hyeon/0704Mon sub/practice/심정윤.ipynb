## CSV 파일 읽고 컬럼별 최고값과 값의 갯수 출력

```python
# 모듈 로딩
import pandas as pd

# 파일 관련 변수 선언
DIR_PATH='../Data/'
FILE_NAME=DIR_PATH+'iris2.csv'
```

```python
# 데이터 로딩
#  File --> Dataframe 으로 로딩하기

iris2DF=pd.read_csv(FILE_NAME,header=None)
```

```python
iris2DF
```

```python
iris2DF.info()
```

```python
iris2DF.columns=['SepalLength','SepalWidth','PetalLength','PetalWidth','Name']
iris2DF
```

```python
for i in iris2DF.columns[:-1]:             # 마지막 Names 컬럼에 해당하는 부분 제외하기  
    print(f'{i} => MAX : {iris2DF[i].max()}  /  MIN : {iris2DF[i].min()}')
    print(f'값의 갯수 : {iris2DF[i].value_counts()}')
```

```python
for i in iris2DF.columns[:-1]:
    print(iris2DF[i].describe().loc[['min','max']])
```
