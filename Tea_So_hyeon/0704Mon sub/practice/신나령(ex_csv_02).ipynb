## [실습]
- csv 파일 읽고 컬럼별 최고값과 값의 개수 출력

```python
import pandas as pd

DIR_PATH='../Data/'
FILE_NAME=DIR_PATH+'iris2.csv'
irisDF=pd.read_csv(FILE_NAME,header=None)#이 데이터에 헤더가 없다고 알려줌, sep=','< 기본
irisDF.info()
```

```python
#최대값
reirisDF=irisDF.iloc[:,:4]
reirisDF=pd.DataFrame(reirisDF.max())
reirisDF.columns=['max']
reirisDF
```

```python
#개수
reirisDF=pd.DataFrame([irisDF[0].count(), irisDF[1].count(), irisDF[2].count(), irisDF[3].count()])
reirisDF.columns=['count']
reirisDF
```

```python
# 값의 개수 => value_counts()
irisDF.columns
for i in irisDF.columns:
    print(f'{irisDF[i].value_counts(dropna=False)}\n')
```
